```python
import numpy as np
import pandas as pd
from numpy.random import randn

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip")
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip",
          scatter_kws={'marker':'o', 'color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip",order=4,
          scatter_kws={'marker':'o', 'color':'indianred'},
          line_kws={'linewidth':1,'color':'blue'})
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', fit_reg=False)
```

```python
tips['tip_percentage']=100*(tips['tip']/tips['total_bill']) #Working out the percentage of how much of the bill are tips

tips.head()
```

```python
sns.lmplot(data=tips, x='size', y='tip_percentage')
```

```python
sns.lmplot(data=tips, x='size', y='tip_percentage', x_jitter=.2)
```

```python
sns.lmplot(data=tips, x='size', y='tip_percentage', x_estimator=np.mean)
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip_percentage', hue='sex', markers=['x','o'], x_jitter=.2)
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip_percentage', hue='day')
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip_percentage', lowess=True, line_kws={'color':'black'}, x_jitter=.1)
```

```python
sns.regplot(data=tips, x='total_bill', y='tip_percentage')
```

```python
fig, (axis1, axis2) = plt.subplots(1,2,sharey=True)

sns.regplot(data=tips, x='total_bill', y='tip_percentage', ax=axis1)
sns.violinplot(data=tips,x='size', y='tip_percentage',color='blue',ax=axis2)
```
